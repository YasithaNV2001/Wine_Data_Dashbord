```python
# Load CSV files into a list
# csv_files = ["https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/Australia_Wine_Stats.csv",
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/Chile_Wine_Stats.csv",
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/France_Wine_Stats.csv",
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/Italy_Wine_Stats.csv", 
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/New%20Zealand_Wine_Stats.csv",
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/Portugal_Wine_Stats.csv",
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/Spain_Wine_Stats.csv ",
#              "https://raw.githubusercontent.com/YasithaNV2001/Wine_Data_Dashbord/refs/heads/Data_Preparation/Wine_Stats/USA_Wine_Stats.csv"]  




csv_files = glob.glob("../Wine_Stats/*.csv") 
data_frames = [pd.read_csv(file) for file in csv_files]

# Concatenate into a single DataFrame
wine_df = pd.concat(data_frames, ignore_index=True)

# Display the first few rows
wine_df.head()
```

```python
# Display the first few rows of the DataFrame
print(wine_df.head())

# Display the DataFrame structure (columns and data types)
print(wine_df.info())

# Summary statistics for numerical columns
print(wine_df.describe())
```

```python
# Split the 'Region' column by '/' and create separate columns for 'Country' and 'Region Detail'
wine_df['Country'] = wine_df['Region'].str.split('/').str[0].str.strip()  # First part as 'Country'
wine_df['Country_Region'] = wine_df['Region'].str.split('/').str[1].str.strip()  # Second part as 'Country_Region'

# Display the first few rows to confirm the changes
print(wine_df[['Region', 'Country', 'Country_Region']].head())
print(wine_df)
```

```python
# Remove duplicate rows from the DataFrame
# Verify by checking the shape of the DataFrame before and after
print("Before removing duplicates:", wine_df.shape)
wine_df = wine_df.drop_duplicates()
print("After removing duplicates:", wine_df.shape)
print("There are no duplicates rows")
```

```python
#check for null values
print("\nnull records:")
print(wine_df.isnull().sum())

```

```python
# Calculate and display the percentage of missing values
print("\nmissing values percentage:")
missing_percentage = (wine_df.isnull().sum() / len(wine_df)) * 100
print(missing_percentage)

```

```python
# Since more than 40% of the 'Grapes' column is missing, it will be dropped as it's not critical for analysis
wine_df = wine_df.drop(columns=['Grapes'])
print(wine_df)
```

```python
#The 'Wine style' column has about 10% missing values, so we will drop rows with missing values in this column
wine_df = wine_df.dropna(subset=['Wine style'])
```

```python
# Display the updated DataFrame info
wine_df.info()
```

```python
# Create a new 'index' column starting from 1
wine_df.insert(0, "Index", range(1, len(wine_df) + 1))
wine_df.drop(columns=['Unnamed: 0'])
```

```python
#Display the updated Dataframe
print(wine_df)

```

```python
# Function to identify outliers based on the 1st and 99th percentiles
def find_percentile_outliers(df, columns, lower_percentile=1, upper_percentile=99):
    outlier_rows = {}

    for column in columns:
        # Calculate the lower and upper bounds based on percentiles
        lower_bound = df[column].quantile(lower_percentile / 100)
        upper_bound = df[column].quantile(upper_percentile / 100)
        
        # Find rows where values are outside the bounds
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)].index
        
        # Get the corresponding row numbers for these outliers
        outlier_row_numbers = df.loc[outliers, 'Index'].tolist()
        
        # Store the row numbers of outliers for each column
        outlier_rows[column] = outlier_row_numbers
    
    return outlier_rows
```

```python
#Columns to check for outliers
columns_to_check = ['Rating', 'Price', 'Bold', 'Tannin', 'Sweet', 'Acidic']
```

```python
# Find the rows with outliers in the specified columns
outlier_rows = find_percentile_outliers(wine_df, columns_to_check, lower_percentile=1, upper_percentile=99)
```

```python
# Print the outlier row numbers for each column
for column, rows in outlier_rows.items():
    print(f"Outliers for column '{column}':")
    print(rows)
    print("\n")
```

```python
# Interpretation of the outliers
print("Interpretation:")
print("Outliers in the 'Rating,' 'Price,' 'Tannin,' 'Sweet,' 'Acidic,' and 'Bold' columns in wines are considered valid and should not be" 
      "removed due to domain-specific reasons. These outliers may represent rare but legitimate cases, such as premium products or exceptional"
      " ratings, which are important for accurately capturing the full range of wine characteristics.")
```

```python

# Create new columns for each food pairing
# Explode the list values in the 'Food pairings' column
from sklearn.preprocessing import MultiLabelBinarizer

# Clean up the 'Food pairings' column if it's in string format
wine_df['Food pairings'] = wine_df['Food pairings'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Use MultiLabelBinarizer to create binary columns for each food
mlb = MultiLabelBinarizer()
food_pairing_encoded = mlb.fit_transform(wine_df['Food pairings'].dropna())

# Get the new columns and merge them into the dataframe
food_pairing_columns = pd.DataFrame(food_pairing_encoded, columns=mlb.classes_, index=wine_df['Food pairings'].dropna().index)
wine_df = pd.concat([wine_df, food_pairing_columns], axis=1)

# Fill missing values for the new columns with False (unpaired foods)
for column in mlb.classes_:
    wine_df[column] = wine_df[column].fillna(False).astype(bool)

# Display the updated DataFrame
print(wine_df.head())

# Save the updated DataFrame to a CSV file
output_path = "data/processed/process_wine_data_with_food_pairings.csv"
wine_df.to_csv(output_path, index=False)
```
